```python
df['FLAG'].fillna('No Flag', inplace=True)
df['FLAG'] = df['FLAG'].replace({'High': 'High Risk', 'Low': 'Low Risk'})
invalid_flags = ['Invalid', 'Not Applicable']
df = df[~df['FLAG'].isin(invalid_flags)]
```

#### 20) Filter the dataset to include only the relevant columns ('INDICATOR', 'UNIT', 'YEAR', 'ESTIMATE', 'STUB_NAME').

```python
filtered_df = df[['INDICATOR', 'UNIT', 'YEAR', 'ESTIMATE', 'STUB_NAME']].copy()  # Make a copy to avoid warnings

# Create a new column to categorize areas as urban or rural
def categorize_area(stub_name):
    if 'Urban' in stub_name:
        return 'Urban'
    elif 'Rural' in stub_name:
        return 'Rural'
    else:
        return 'Other'

filtered_df.loc[:, 'Area_Type'] = filtered_df['STUB_NAME'].apply(categorize_area)  # Use .loc to set values

# Calculate summary statistics
summary_stats = filtered_df.groupby('Area_Type')['ESTIMATE'].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Area_Type', y='ESTIMATE', data=filtered_df)
plt.title('Distribution of Drug Overdose Death Rates between Urban and Rural Areas')
plt.xlabel('Area Type')
plt.ylabel('Death Rate per 100,000 Population')
plt.show()

# Display summary statistics
print("Summary Statistics for Drug Overdose Death Rates by Area Type:")
print(summary_stats)
```
